```
import io
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
print("\t \t \t Data Pusat","\t \t \t Data Daerah")
data = pd.read_excel(io.BytesIO(uploaded['Kasus Per-Hari Provinsi Jawa Barat.xlsx']))
data
```

```
data.describe()
```

```
DataPusat = [data[ 'Positif'],data['Sembuh'],data['Meninggal']]
DataDaerah = [data[ 'Positif.1'],data['Sembuh.1'],data['Meninggal.1']]
print("Total Positif Data Pusat = ",DataPusat[0].sum())
print("Total Positif Data Daerah = ",DataDaerah[0].sum())
print("Total Sembuh Data Pusat = ",DataPusat[1].sum())
print("Total Sembuh Data Daerah = ",DataDaerah[1].sum())
print("Total Meninggal Data Pusat = ",DataPusat[2].sum())
print("Total Meninggal Data Daerah = ",DataDaerah[2].sum())
print("Selisih Antara Positif Data Pusat dan Positif Data Daerah = ", abs(DataPusat[0].sum() - DataDaerah[0].sum()))
print("Selisih Antara Sembuh Data Pusat dan Sembuh Data Daerah = ", abs(DataPusat[1].sum() - DataDaerah[1].sum()))
print("Selisih Antara Meninggal Data Pusat dan Meninggal Data Daerah = ", abs(DataPusat[2].sum() - DataDaerah[2].sum()))
```

```
plt.figure(figsize=(20,10))
x1 = data['Tanggal']
y1 = data['Positif']
x2 = data['Tanggal']
y2 = data['Positif.1']
plt.plot(x1, y1,'r', label = "Positif Data Pusat",)
plt.plot(x2, y2,'b', label = "Positif Data Daerah")
plt.xlabel('Bulan')
plt.ylabel('Positif')
plt.legend()
plt.title("Diagram Garis Kasus Positif Per Hari")
plt.show()
```

```
plt.figure(figsize=(20,10))
x1 = data['Tanggal']
y1 = data['Sembuh']
x2 = data['Tanggal']
y2 = data['Sembuh.1']
plt.plot(x1, y1,'r', label = "Sembuh Data Pusat",)
plt.plot(x2, y2,'b', label = "Sembuh Data Daerah")
plt.xlabel('Bulan')
plt.ylabel('Sembuh')
plt.legend()
plt.title("Diagram Garis Kasus Sembuh Per Hari")
plt.show()
```

```
plt.figure(figsize=(20,10))
x1 = data['Tanggal']
y1 = data['Meninggal']
x2 = data['Tanggal']
y2 = data['Meninggal.1']
plt.plot(x1, y1,'r', label = "Meninggal Data Pusat",)
plt.plot(x2, y2,'b', label = "Meninggal Data Daerah")
plt.xlabel('Bulan')
plt.ylabel('Meninggal')
plt.legend()
plt.title("Diagram Garis Kasus Meninggal Per Hari")
plt.show()
```

Berdasarkan hasil yang didapat. Dapat disimpulkan bahwa adanya sedikit perbedaan antara data dari pusat dan data dari pemda jawa barat sendiri.Terutama di jumlah orang yang meninggal karena covid-19, terlihat perbedaan yang cukup jelas pada sekitar pertengahan sampai akhir bulan 10. Namun untuk kasus positif dan sembuhnya sendiri, perbedaannya hanya terdapat di beberapa titik dan tidak ada perbedaan yang signifikan.

Namun, walaupun perbedaan yang terlihat jelas pada grafik adalah kasus meninggal, ternyata selisih kasus sembuh antara data pusat dan data daerah lebib besar daripada kasus meninggal, yaitu sebesar 211 kasus. Sementara selisih kasus meninggal adalah 135 kasus.
